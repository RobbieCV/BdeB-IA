```python
import numpy as np
import pandas as pd

data = pd.read_csv('/Users/Dominique/Documents/BIOptilab/CEGEP/CEGEP_BDB/IA/DATASETS/RULES/market_basket.csv',delimiter=";")
```

```python
data.head()
```

```python
# transformer en tableau binaire
TC=pd.crosstab(data.ID,data.Product)
TC
```

```python
from mlxtend.frequent_patterns import apriori
```

```python
freq_itemsets = apriori(TC,min_support=0.03,max_len=3,use_colnames=True)
```

```python
print(freq_itemsets)
```

```python
def is_inclus(x,items):
    return items.issubset(x)

id = np.where(freq_itemsets.itemsets.apply(is_inclus,items={"Potatoes"}))
print(id)
```

```python
print(freq_itemsets.loc[id])
```

```python
from mlxtend.frequent_patterns import association_rules
```

```python
regles = association_rules(freq_itemsets,metric="confidence",min_threshold=0.80)
```

```python
print(regles.iloc[:100,0:7])
```
